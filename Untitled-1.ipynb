```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Step 1: Data Cleaning
This involves handling missing values, incorrect formats, and outliers. For this example, we'll focus on checking for missing values and ensuring the data types are correct.


```python
import pandas as pd

# Load the dataset
sales_data = pd.read_csv('C:/Users/Rohan/Downloads/sales_data_large.csv')

# Display the first few rows of the dataset
print(sales_data.head())

# Check for missing values
missing_values = sales_data.isnull().sum()
print("Missing Values:\n", missing_values)

# Check data types
print("Data Types:\n", sales_data.dtypes)

# Assuming there are no missing values, we can proceed. If there were, we might fill or drop them.
```

Step 2: Data Aggregation
Aggregate the sales data at different levels, such as by date, store, and product.

```python
daily_sales = sales_data.groupby('Date').agg(
    total_units_sold=('Units Sold', 'sum'),
    total_sales_amount=('Sales Amount', 'sum')
).reset_index()

# Aggregate sales data by store
store_sales = sales_data.groupby('Store ID').agg(
    total_units_sold=('Units Sold', 'sum'),
    total_sales_amount=('Sales Amount', 'sum')
).reset_index()

# Aggregate sales data by product
product_sales = sales_data.groupby('Product ID').agg(
    total_units_sold=('Units Sold', 'sum'),
    total_sales_amount=('Sales Amount', 'sum')
).reset_index()

print("Daily Sales:\n", daily_sales)
print("Store Sales:\n", store_sales)
print("Product Sales:\n", product_sales)
```

Step 3: Analysis
Calculate total sales and average sales per product.
Identify the store with the highest sales performance.
Analyze the impact of discounts on sales amounts.

```python
# Total sales and average sales per product
total_sales_per_product = product_sales['total_sales_amount'].sum()
average_sales_per_product = product_sales['total_sales_amount'].mean()

# Store with the highest sales performance
best_store = store_sales.loc[store_sales['total_sales_amount'].idxmax()]

# Analyze the impact of discounts on sales amounts
# We can calculate the correlation between discount and sales amount
correlation_discount_sales = sales_data[['Discount Applied', 'Sales Amount']].corr().iloc[0, 1]

print("Total Sales per Product: ", total_sales_per_product)
print("Average Sales per Product: ", average_sales_per_product)
print("Best Store:\n", best_store)
print("Correlation between Discount and Sales Amount: ", correlation_discount_sales)
```

Step 4: Visualization
Using Matplotlib to visualize the results.

```python
import matplotlib.pyplot as plt

# Sales trends over time
plt.figure(figsize=(12, 6))
plt.plot(daily_sales['Date'], daily_sales['total_sales_amount'], marker='o')
plt.title('Sales Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()

# Sales distribution across different stores
plt.figure(figsize=(12, 6))
plt.bar(store_sales['Store ID'], store_sales['total_sales_amount'], color='skyblue')
plt.title('Sales Distribution Across Different Stores')
plt.xlabel('Store ID')
plt.ylabel('Total Sales Amount')
plt.grid()
plt.tight_layout()
plt.show()

# Performance comparison of products
plt.figure(figsize=(12, 6))
plt.bar(product_sales['Product ID'], product_sales['total_sales_amount'], color='lightgreen')
plt.title('Performance Comparison of Products')
plt.xlabel('Product ID')
plt.ylabel('Total Sales Amount')
plt.grid()
plt.tight_layout()
plt.show()
```
